```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
df = pd.read_csv('Groceries_dataset.csv')
```

```python
print(len(df))
```

```python
print(df[['Member_number','itemDescription']])
```

```python
df.head()
```

```python
baskets = df.groupby('Member_number')['itemDescription'].apply(list).values.tolist()
print(len(baskets))
```

```python
te = TransactionEncoder()
te_ary = te.fit(baskets).transform(baskets)
df_encoded = pd.DataFrame(te_ary, columns=te.columns_)

print(df_encoded.head())
```

```python

frequent_itemsets = apriori(df_encoded, min_support=0.01, use_colnames=True)
frequent_itemsets = frequent_itemsets.sort_values(by='support', ascending=False)

print(frequent_itemsets.head())
```

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)

clean_rules = rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']]


print(clean_rules.to_string(index=False))
```

```python
top_rule = clean_rules.iloc[0]
if_item = list(top_rule['antecedents'])[0]
then_item = list(top_rule['consequents'])[0]
conf = top_rule['confidence'] * 100
lift_val = top_rule['lift']

print(f"Rule Example: If a customer buys -> {if_item}, they will likely buy -> {then_item}.")
print(f" • Confidence ({conf:.1f}%): Out of all customers who bought {if_item}, {conf:.1f}% also bought {then_item}.")
print(f" • Lift ({lift_val:.2f}): A Lift greater than 1 means {then_item} is strongly boosted by the purchase of {if_item}.")
```
